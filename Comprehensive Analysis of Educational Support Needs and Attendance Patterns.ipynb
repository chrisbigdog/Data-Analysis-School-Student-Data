1: Load and Explore the Data:<br>
1.1 – Load the Data

```python
import pandas as pd

# Load the Excel file
file_path = 'CLEANED NonIdentifiable Data T1 2023.xlsx'
data = pd.read_excel(file_path, sheet_name='Cleaned_NonIdentifiable_Data_20')

# Display the first few rows
data.head()
```

1.2 – Explore Basic Information
Understanding the structure of the data:

```python
# Basic information about the data
data.info()

# Summary statistics (for all columns)
data.describe(include='all')
```

1.3 – Check for Missing Values

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

2: Data Cleaning and Preprocessing:<br>
Goal: Ensure the data is consistent, free of errors, and ready for analysis.
We will handle missing values, remove duplicates, and standardize categorical data.<br> 2.1 – Handle Missing Values - address any missing values.

```python
# Option 1: Fill missing values with a placeholder (e.g., 'Unknown')
data.fillna('Unknown', inplace=True)

# Option 2: Drop rows with missing values (if appropriate)
# data.dropna(inplace=True)

# Verify missing values are handled
print("Missing Values After Handling:\n", data.isnull().sum())
```

2.2 – Remove Duplicates:
Ensure no duplicate rows exist:

```python
# Remove duplicate rows (if any)
data.drop_duplicates(inplace=True)

# Verify duplicates are removed
print("Remaining Duplicates:\n", data.duplicated().sum())
```

2.3 – Standardize Categorical Data:
Check for inconsistencies in categorical columns:

```python
# Example: View unique values in 'Gender' column
print("Unique Values in Gender:\n", data['Gender'].unique())

# Standardize (if needed)
data['Gender'] = data['Gender'].replace({'F': 'Female', 'M': 'Male'})
```

2.4 – Rename Columns for Clarity (Optional, but will do it anyway)

```python
# Rename columns to remove extra spaces or special characters
data.columns = data.columns.str.strip().str.replace('\n', '_').str.replace(' ', '_')

# Verify updated column names
print("Updated Columns:\n", data.columns)
```

2.5 – Save the Cleaned Data (Optional)

```python
# Save the cleaned data for further use
data.to_csv('cleaned_data.csv', index=False)
```

3: Visualize Patterns and Trends:
In this step, we'll create visualizations to identify patterns in gender distribution, attendance, and areas of need.<br>
3.1 – Import Visualization Libraries: Importing the necessary libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot styles
sns.set(style='whitegrid')
```

3.2 – Gender Distribution:
Visualize how the data is distributed by gender:

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.show()
```

Insight: This reveals if there's a gender imbalance in the dataset.

3.3 – Attendance by Year Level: Analyze attendance rates across different year levels:

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Year_Level', hue='Attendance', data=data)
plt.title('Attendance by Year Level')
plt.xticks(rotation=45)
plt.show()
```

Insight: This shows attendance patterns across year levels, helping identify trends in absenteeism.

3.4 – Area of Need Breakdown: Understand the main areas where support is required:

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Area_of_Need', data=data, order=data['Area_of_Need'].value_counts().index)
plt.title('Area of Need Breakdown')
plt.show()
```

Insight: This highlights the most common areas where students need support (e.g., learning, behavioral, physical).

3.5 – Priority of Support: Visualize how support needs are prioritized:

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Priority_of__Support', data=data)
plt.title('Priority of Support Levels')
plt.show()
```

Insight: This indicates whether most students require urgent or non-urgent support.

3.6 – Attendance by Ethnicity

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Ethnicity', hue='Attendance', data=data)
plt.title('Attendance by Ethnicity')
plt.xticks(rotation=45)
plt.show()
```

Insight: This can reveal trends in attendance rates across different ethnic groups.

4: Identify Outliers and Anomalies: In this step, we'll look for unusual patterns or anomalies that may indicate areas requiring further investigation. Since most of the data is categorical, we'll focus on identifying inconsistencies or rare categories.<br>
4.1 – Check for Rare Categories: Identify categories that appear infrequently:

```python
# Check for rare values in key categorical columns
for column in ['Attendance', 'Area_of_Need', 'Priority_of__Support']:
    print(f"\nValue Counts for {column}:")
    print(data[column].value_counts())
```

Insight: This helps to highlight categories that occur less frequently, which could be outliers or special cases.

4.2 – Box Plot for Numeric Outliers: If the dataset includes numerical values, use box plots to identify outliers:

```python
# Check for numeric columns
numeric_columns = data.select_dtypes(include=['number']).columns

# Plot boxplots for numeric columns
if len(numeric_columns) > 0:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=data[numeric_columns])
    plt.title('Boxplot for Numeric Columns')
    plt.show()
else:
    print("No numerical columns to visualize.")
```

Insight: Box plots highlight data points that fall far outside the interquartile range (potential outliers).

4.3 – Visualize Unusual Patterns: Analyze if specific groups (e.g., certain year levels or ethnicities) exhibit unusual attendance patterns:

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Year_Level', hue='Attendance', data=data)
plt.title('Outliers in Attendance by Year Level')
plt.xticks(rotation=45)
plt.show()
```

Insight: This can reveal attendance anomalies across different year levels.

4.4 – Cross-Check Anomalies with Support Needs:

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Area_of_Need', hue='Priority_of__Support', data=data)
plt.title('Support Prioritization by Area of Need')
plt.show()
```

Insight: Look for areas where "Urgent" support appears unexpectedly in typically low-priority areas.

4.5 – Filter Potential Anomalies: Isolate rare or concerning patterns:

```python
# Example: Filter students with urgent support needs but regular attendance
anomalies = data[(data['Priority_of__Support'] == 'Urgent') & 
                 (data['Attendance'] == 'Regular attendance 90%-100%')]

print("Potential Anomalies:\n", anomalies)
```

Insight: This narrows down unusual cases for further review.

The anomalies found highlight cases where students with urgent support needs are still marked as having regular attendance (90%-100%). This could indicate potential gaps in identifying the full extent of their challenges or highlight situations where students maintain regular attendance despite significant support needs.

- Gaps in identifying the full extent of their challenges.

### 2. Diverse Areas of Need

- **Emotional and Behavioral** issues (e.g., managing emotions, anxiety).

- **Internal** providers handle most cases, but **other providers** are occasionally involved.  
- **Monitoring vs. Active** response statuses show variability in follow-up intensity.  

---

## Interpretation  
### 1. Mismatch Between Need and Attendance  
- Students flagged for urgent intervention continue attending regularly. This could indicate:  
  - **Conditions are manageable** but flagged early to prevent deterioration.  
  - **Emotional or behavioral needs** that may be masked during school hours.  
  - Urgency might be tied to factors beyond attendance (e.g., emotional health).  

### 2. Potential Gaps in Support  
- **Emotional regulation and medical needs** may require **external assessments** or additional resources.  
- Cases lacking **specific response details** may need further documentation and monitoring.  

---

## Next Steps for Actionable Insights  
### 1. Follow-Up  
- Review individual cases to understand why urgent support is flagged despite regular attendance.  
- Schedule follow-ups with **teachers or caregivers** to monitor ongoing progress.  

### 2. Resource Allocation  
- Prioritize **in-class adaptations** or **mental health resources** for emotional or physical needs.  
- Consider **external evaluations** for students with "Not Provided" under diagnosis.  

### 3. Policy Adjustment  
- Develop clearer criteria for urgent support cases, ensuring attendance alone isn’t the primary indicator of well-being.  
- Implement **preventive measures** for students with emotional or behavioral needs to avoid escalation.

Create Interactive Dashboard: Interactive dashboard to visualize gender distribution, support needs, and anomalies.

```python
import dash
from dash import dcc, html, Input, Output
import plotly.express as px
import pandas as pd

# Load the data
file_path = 'CLEANED NonIdentifiable Data T1 2023.xlsx'
data = pd.read_excel(file_path, sheet_name='Cleaned_NonIdentifiable_Data_20')
data.columns = data.columns.str.replace('\n', '_').str.strip()

# Prepare the anomaly data
anomalies = data[(data['Priority_of__Support'] == 'Urgent') & 
                 (data['Attendance'] == 'Regular attendance 90%-100%')]

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the dashboard
app.layout = html.Div([
    html.H1("Educational Support Needs Dashboard", style={'textAlign': 'center', 'color': 'green'}),

    # Gender Distribution Pie Chart
    html.Div([
        dcc.Graph(
            id='gender-pie-chart',
            figure=px.pie(data, names='Gender', title='Gender Distribution')
        )
    ]),

    # Attendance by Year Level
    html.Div([
        dcc.Graph(
            id='attendance-bar-chart',
            figure=px.bar(data, x='Year Level', color='Attendance', title='Attendance by Year Level')
        )
    ]),

    # Priority of Support by Area of Need
    html.Div([
        dcc.Graph(
            id='area-need-bar-chart',
            figure=px.bar(data, y='Area_of_Need', color='Priority_of__Support',
                          title='Priority of Support by Area of Need')
        )
    ]),

    # Anomalies Bar Chart
    html.Div([
        dcc.Graph(
            id='anomaly-bar-chart',
            figure=px.bar(anomalies, x='Year Level', color='Priority_of__Support',
                          title='Anomalies in Urgent Support by Year Level')
        )
    ])
])

# Run the dashboard
if __name__ == '__main__':
    app.run_server(debug=True)
```
